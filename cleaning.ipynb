```python
# load data
import pandas as pd 
df=pd.read_excel("/drive/sales_raw_500.xlsx")

df
```

```python
#remove duplicates
df=df.drop_duplicates()
```

```python
df
```

```python
#fix date coloumn
df['Order_Date']=pd.to_datetime(df['Order_Date'],errors='coerce')
```

```python
df=df.dropna(subset=['Order_Date'])
```

```python
df
```

```python
# diffrent coloumn profit year and month from order date
df = df.copy()
df['Profit'] = df['Sales'] - df['Cost']
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
```

```python
df
```

```python
#to save file in csv
df.to_csv('sales_cleaned.csv',index=False)
```
